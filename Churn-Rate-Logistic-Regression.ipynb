```python
#X=df.drop(['StreamingMovies','OnlineBackup','Contract','Churn'],1)

X=df.drop(['Churn'],1)
```

y=df['Churn']

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=123)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
lr=LogisticRegression()
```

```python
model=lr.fit(X_train,y_train)
```

```python
model.coef_
```

```python
classes=model.predict(X_test)
```

```python
#classes.size
```

```python
print("Model Accuracy")
metrics.accuracy_score(y_test,classes)
```

```python
print("Classification Report")
metrics.classification_report(y_test,classes)
```

```python
print("Roc Auc")
metrics.roc_auc_score(y_test,classes)
```

```python
# ROC chart
```

```python
fpr,tpr,th=roc_curve(y_test,classes)
roc_auc=metrics.auc(fpr,tpr)
import matplotlib.pyplot as plt
plt.title("Roc Chart")
plt.plot(fpr,tpr,'b',label='AUC=% 0.2f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'o--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel("True Posative rate")
plt.xlabel("False Posative rate")
plt.show()
```

```python
# Confusion Matrix
```

```python
print("Confisusion Matrix")
cf=metrics.confusion_matrix(y_test,classes)
lbl1=["Predicted 0","Predicted1"]
lbl2=["True 0","True 1"]
sns.heatmap(cf,annot=True,cmap="Greens",fmt='d',xticklabels=lbl1,yticklabels=lbl2)
plt.show()
```
